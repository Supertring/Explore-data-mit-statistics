__Normal Distributions__

* A probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurance than data far from the mean, and is also know as the Gaussian distribution
* It appears as a bell curve

* Based on two parameters:
    * Mean: $\mu$ = $\sum_{i=1}^{n} \frac{x_i}{n}$
        * Indicates the `central` or `average value` of the entire dataset
    * Standard Deviation: $\sigma$ : $\sqrt{\frac{\sum ({x_i - \mu})^2}{N}}$
        * Indicates the `spread` or `variations` of data points around the `mean`

```python
from scipy.stats import norm
```

```python
# generate data between -3 to +3
x_axis = np.arange(-3,3,0.1)
x_axis
```

```python
# mean=0, sd=1.
y_axis = norm.pdf(x_axis, 0,1)
y_axis
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

## Cumulative Distribution Function (CDF)

* CDF is a method to describe the distribution of random variables.
* The advantage of CDF is that is can be define for any kind of random variable (discrete, continuous, and mixed)

__Definition__:

CDF of a random variable $X$ is defined as:

$F_{X}(x) = P(X \le x), for all x \in \mathbb{R}$

Subscript $X$: indicates this is a CDF of the random variable $X$.

```python
import numpy as np
import matplotlib.pyplot as plt

#given dataset
data = [0.5, 0.6, 0.1, 0.5, 0.7, 0.8, 0.9]

#sort the data in ascending order
data_sorted = np.sort(data)
data_sorted
```

```python
#calculate the cdf
cdf_values = np.arange(1, len(data_sorted) +1) / len(data_sorted)
```

```python
cdf_values
```

```python
plt.step(data_sorted, cdf_values, where='post')
plt.xlabel('sorted data')
plt.ylabel('cdf values')
plt.title('CDF')
plt.grid()
plt.show()
```

```python
scipy.stats.norm.cdf(data_sorted)
```

```python
from scipy.stats import norm
```

```python
dmean = round(data_sorted.mean(),3)
sd = round(data_sorted.std(),3)
norm.cdf(data_sorted, loc=dmean, scale=sd )
```
